```python
import pandas as pd
import matplotlib.pyplot as plt

# importando os dois datasets para a classe DataFrame do Pandas

df_quiz = pd.read_csv('./case estágio de dados/quiz-sample.csv')
df_orders = pd.read_csv('./case estágio de dados/orders-sample.csv')

df_orders['cancelled_at'].isnull().value_counts()
```

```python
# criando um novo DataFrame com os dados da person_id relacionados de uma tabela na outra

df_merge = pd.merge(df_quiz, df_orders, on=['person_id'], how='inner').drop_duplicates(['person_id', 'preocupacao_de_pele']).reset_index()

# merge -> faz a unificação dos dados que possuem relação, neste caso, person_id dos dois DataFrames
# drop_duplicates -> para remover pessoas duplicadas do DataFrame (existiam pessoas que tinha uma mesma preocupação, então retirei
#                    as que possuiam essas duplicações)
# reset_index -> para colocar os index em ordem

df_merge
```

```python
# ver quantas preocupações existiam no DataFrame original

preocupacoes = df_quiz[['preocupacao_de_pele']].drop_duplicates().sort_values('preocupacao_de_pele').set_index('preocupacao_de_pele')

preocupacoes
```

```python
# contar quantas preocupações existem em cada uma das suas classes

# df_merge.groupby(['preocupacao_de_pele']).count()

quantidade = df_merge['preocupacao_de_pele'].value_counts()

quantidade
```

```python
# verificação por "força bruta", para verificar se o resultado bate com o de cima

cont = 0

for i in range(len(df_merge)):
    try:
        if(df_merge['preocupacao_de_pele'][i] == 'pp_oleosidade'):
            cont += 1
    except:
        pass

cont
```
