```python
import pandas as pd

# importando os dois datasets para a classe DataFrame do Pandas

df_quiz = pd.read_csv('./case estágio de dados/quiz-sample.csv', parse_dates=['data_submissao'])
df_orders = pd.read_csv('./case estágio de dados/orders-sample.csv', parse_dates=['processed_at', 'cancelled_at'])

df_quiz
```

```python
df_orders
```

```python
df_quiz.groupby([df_quiz['data_submissao'].dt.year, df_quiz['data_submissao'].dt.month, 'preocupacao_de_pele'])[['preocupacao_de_pele']].count().rename(columns={'preocupacao_de_pele': 'Total'})
```

```python
# ver quantas preocupações existiam no DataFrame original

preocupacoes = df_quiz[['preocupacao_de_pele']].drop_duplicates().sort_values('preocupacao_de_pele').set_index('preocupacao_de_pele')

preocupacoes
```

```python
# criando um novo DataFrame com os dados da person_id relacionados de uma tabela na outra

df_merge = pd.merge(df_quiz, df_orders, on=['person_id'], how='inner') # .drop_duplicates(['person_id', 'preocupacao_de_pele'])

# merge -> faz a unificação dos dados que possuem relação, neste caso, person_id dos dois DataFrames
# drop_duplicates -> para remover pessoas duplicadas do DataFrame (existiam pessoas que tinha uma mesma preocupação, então retirei
#                    as que possuiam essas duplicações)
# reset_index -> para colocar os index em ordem

df_merge
```

```python
df_merge.groupby(df_merge['data_submissao'].dt.year).size().reset_index().set_index('data_submissao')
```

```python
df_merge.groupby([df_merge['data_submissao'].dat.year, 'preocupacao_de_pele'])[['preocupacao_de_pele']].count().rename(columns={'preocupacao_de_pele': 'Total'})
```

```python
# agrupa por ano e mês e conta quantas preocupações existem em cada mês

df_quant = df_merge.groupby([df_merge['data_submissao'].dt.year, df_merge['data_submissao'].dt.month, 'preocupacao_de_pele'])[['preocupacao_de_pele']].count().rename(columns={'preocupacao_de_pele': 'Total'})

df_quant
```

### 2. Observando como as preocupações e seus pesos relativos se movimentam ao longo do período analisado na resposta do item 1:
- #### Quais são as hipóteses que poderiam justificar esses movimentos?

Pode-se observar nas variações mês a mês que existe uma sazionalidade das preocupações, onde as preocupações com a pele aumentam em determinados períodos do ano, sendo o principal período durante os meses 4 ao mês 9. A hipótese para este movimento de sazionalidade pode ser explicada pela temperatura e pelo clima predominante nesta estação do ano, pois o tempo seco e as temperaturas baixas podem aumentar as preocupações existentes com a pele, como o ressecamento da pele e consequentemente trazer outras preocupações. Desta forma, a temperatura e o clima da estação podem influenciar nas preocupações da pele.

De tal modo, em 2020 também tivemos da pandemia do COVID-19, onde muitas pessoas tiveram que ficar em casa. Pode-se observar assim, que no começo da pandemia existiu um grande aumento das preocupações com a pele, podendo ser influenciado por causa que as pessoas precisaram ficar em casa isoladas e, analisando esse fato, pode-se dizer que tiveram mais tempo para cuidar da saúde de suas peles. Assim, com a análise mais cuidadosa de suas peles, preocupações recorrentes ou não podem aparecer e assim influenciar em suas submissões. Desta forma, pode-se concluir que durante a pandemia existiu mais um movimento de preocupação maior com a pele.

Também analisando o investimento na campanha de marketing em cada produto, pode-se perceber que produtos que tiveram maior investimento obtiveram maior quantidade de compras, desta forma, é verídico qu existe um movimento relacionado à quantidade de investimento feita em determinados produtos, desta forma, quanto mais investimentos em determinados produtos, mais compras podem existir deste produto. Algumas hipóteses seriam mais difíceis de validar devido a complexidade de seus dados, como os movimentos da pandemia, por exemplo, são dados que já foram e teria que ser realizadas pesquisas para validá-los e provavelmente não seria algo que traria alguma clareza ou, talvez, seria eficiente caso alguma pesquisa de consumo aborde esse tema de ter um tempo para cuidar mais da pele, então talvez seria interessante validar para comprovar esse movimento para poder tirar algum lucro sobre essa hipótese.
