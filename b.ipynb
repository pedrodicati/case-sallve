```python
import pandas as pd
import matplotlib.pyplot as plt

# importando os dois datasets para a classe DataFrame do Pandas

df_quiz = pd.read_csv('./case estágio de dados/quiz-sample.csv', parse_dates=['data_submissao'])
df_orders = pd.read_csv('./case estágio de dados/orders-sample.csv', parse_dates=['processed_at', 'cancelled_at'])

df_quiz
```

```python
df_quiz.groupby([df_quiz['data_submissao'].dt.year, df_quiz['data_submissao'].dt.month, 'preocupacao_de_pele'])[['preocupacao_de_pele']].count().rename(columns={'preocupacao_de_pele': 'Total'})
```

```python
# ver quantas preocupações existiam no DataFrame original

preocupacoes = df_quiz[['preocupacao_de_pele']].drop_duplicates().sort_values('preocupacao_de_pele').set_index('preocupacao_de_pele')

preocupacoes
```

```python
# criando um novo DataFrame com os dados da person_id relacionados de uma tabela na outra

df_merge = pd.merge(df_quiz, df_orders, on=['person_id'], how='inner') # .drop_duplicates(['person_id', 'preocupacao_de_pele'])

# merge -> faz a unificação dos dados que possuem relação, neste caso, person_id dos dois DataFrames
# drop_duplicates -> para remover pessoas duplicadas do DataFrame (existiam pessoas que tinha uma mesma preocupação, então retirei
#                    as que possuiam essas duplicações)
# reset_index -> para colocar os index em ordem

df_merge
```

```python
df_merge.groupby(df_merge['data_submissao'].dt.year).size().reset_index().set_index('data_submissao')
```

```python
df_merge.groupby([df_merge['data_submissao'].dt.year, 'preocupacao_de_pele'])[['preocupacao_de_pele']].count().rename(columns={'preocupacao_de_pele': 'Total'})
```

```python
# agrupa por ano e mês e conta quantas preocupações existem em cada mês

df_quant = df_merge.groupby([df_merge['data_submissao'].dt.year, df_merge['data_submissao'].dt.month, 'preocupacao_de_pele'])[['preocupacao_de_pele']].count().rename(columns={'preocupacao_de_pele': 'Total'})

df_quant
```

```python
# exporta para um arquivo para visualizar inteiro

df_quant.to_string('output.txt')
```
