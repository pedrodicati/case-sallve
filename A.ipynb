```python
import pandas as pd

df_facebook = pd.read_csv('./case estágio de dados/last-30d-facebook.csv')

df_facebook
```

```python
# função para converter para int o valor em R$

def delete_real(data: pd.Series) -> pd.DataFrame:
    delete = "R$ ."

    for i in range(len(data)):
        for x in delete:
            if x in data[i]:
                data[i] = data[i].replace(x, "")
    
    data = pd.DataFrame((data.astype(int)))

    return data

df_gasto = delete_real(df_facebook['Gasto']).to_dict()
df_invest = delete_real(df_facebook['Investimento alocado']).to_dict()
df_receita = delete_real(df_facebook['Receita']).to_dict()

# cria um novo DataFrame com os dados convertidos para facilitar a análise e os cálculos

df_total = pd.DataFrame(data=(df_gasto["Gasto"], df_invest['Investimento alocado'], df_receita['Receita'])).transpose().rename(columns={0: 'Gasto', 1: 'Investimento alocado', 2: 'Receita'})
```

```python
list_result = []    # lista para armazenar os resultados de cada campanha em ordem

# laço para calcular a relação de gasto e receita em cada uma das campanhas

for i in range(len(df_total)):
    porcentagem = (df_total['Receita'][i] * 100) / df_total['Gasto'][i]

    list_result.append(f"{porcentagem:.2f} %")

list_result
```

```python
# adiciona uma nova linha da relação de gasto e receita no dataframe original

df_facebook['Relação Gasto e Receita'] = list_result

df_facebook
```
